```python
import pandas as pd
import re

Action_df= pd.read_csv('Action_Movies.csv')
Adventure_df= pd.read_csv('Adventure_Movies.csv')
Comedy_df= pd.read_csv('Comedy_Movies.csv')
Crime_df= pd.read_csv('Crime_Movies.csv')
Documentary_df= pd.read_csv('Documentary_Movies.csv')
Drama_df= pd.read_csv('Drama_Movies.csv')
Horror_df= pd.read_csv('Horror_Movies.csv')
Mystery_df= pd.read_csv('Mystery_Movies.csv')
Romance_df= pd.read_csv('Romance_Movies.csv')
Thriller_df= pd.read_csv('Thriller_Movies.csv')

Movies_combined = pd.concat([Action_df, Adventure_df, Comedy_df, Crime_df, Documentary_df, Drama_df, Horror_df, 
                              Mystery_df, Romance_df, Thriller_df], ignore_index=True)
```

```python
def Title_changes(T):
    if isinstance(T, str):
        T= re.sub(r'^\d+\.\s*', '', T) # Remove numbers and dots
    return T

Movies_combined['Title'] = Movies_combined['Title'].apply(Title_changes)

def K_Conversion(v):
    if isinstance(v, str):
        # Remove parentheses (240k) to 240k
        v = re.sub(r'^\((.*)\)$', r'\1', v.strip())

        # Check if 'K' is present
        if 'K' in v:
            v = v.replace('K', '')
            return round(float(v) * 1000)
    return v

Movies_combined['Votings'] = Movies_combined['Votings'].apply(K_Conversion).astype(int)

def duration_to_minutes(duration):
    if isinstance(duration, str):
        match = re.match(r'(?:(\d+)h)?\s*(?:(\d+)m)?', duration)
        if match:
            hours = int(match.group(1)) if match.group(1) else 0
            minutes = int(match.group(2)) if match.group(2) else 0
            dur= hours * 60 + minutes
    return dur

Movies_combined['Duration'] = Movies_combined['Duration'].apply(duration_to_minutes)

Movies_combined.to_csv('IMDB_movie_list_Converted.csv', index=False)
```
